## BOSTON

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
boston= pd.read_csv("boston_housing.csv")
```

```python
boston.dropna(inplace=True)
boston.info()
```

```python
boston_df=boston[["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]]
boston_df
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(boston_df[["CRIM","ZN","INDUS","NOX","RM","AGE","DIS","PTRATIO","B","LSTAT","MEDV"]]) 
boston_data_normalized = scaler.transform(boston_df[["CRIM","ZN","INDUS","NOX","RM","AGE","DIS","PTRATIO","B","LSTAT","MEDV"]])# Scaled data 
boston_data_normalized
```

```python
boston_data_normalized=pd.DataFrame(boston_df[["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"]])
boston_data_normalized=boston_data_normalized.join(boston_df[boston_df.columns.drop(["CRIM","ZN","INDUS","CHAS","NOX","RM","AGE","DIS","RAD","TAX","PTRATIO","B","LSTAT","MEDV"])])
boston_data_normalized
```

```python
Y = boston_data_normalized['MEDV']
X = boston_data_normalized[boston_data_normalized.columns.drop('MEDV')]
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2, random_state=100)
X_train.shape , X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error
train_predictions = model.predict(X_train)
train_RMSE=mean_squared_error(Y_train,train_predictions)**100
test_predictions = model.predict(X_test)
test_RMSE=mean_squared_error(Y_test,test_predictions)**100
print('Train RMSE : ',train_RMSE)
print('Test RMSE : ',test_RMSE)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
yield_ = pd.read_csv("yield_df.csv")
```

```python
print(yield_.shape)
```

```python
yield_.dropna(inplace=True)
```

```python
yield_.info()
```

```python
yield_df = yield_[["Area","Item","Year","hg/ha_yield","average_rain_fall_mm_per_year","pesticides_tonnes","avg_temp"]]
```

```python
cat_features =[ "Area","Item"]
yield_df= pd.get_dummies(yield_df,columns=cat_features)
print(yield_df.columns)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(yield_df[["Year","hg/ha_yield","average_rain_fall_mm_per_year","pesticides_tonnes","avg_temp"]]) 
yield_df_normalized = scaler.transform(yield_df[["Year","hg/ha_yield","average_rain_fall_mm_per_year","pesticides_tonnes","avg_temp"]])# Scaled data 
yield_df_normalized
```

```python
yield_df_normalized=pd.DataFrame(yield_df,columns=["Year","hg/ha_yield","average_rain_fall_mm_per_year","pesticides_tonnes","avg_temp"])
yield_df_normalized=yield_df_normalized.join(yield_df[yield_df.columns.drop(["Year","hg/ha_yield","average_rain_fall_mm_per_year","pesticides_tonnes","avg_temp"])])
```

```python
Y = yield_df_normalized['hg/ha_yield']
X = yield_df_normalized[yield_df_normalized.columns.drop('hg/ha_yield')]
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1, random_state=100)
X_train.shape , X_test.shape
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train,Y_train)
train_score=model.score(X_train,Y_train)
test_score=model.score(X_test,Y_test)
print('Train Score (R-Squared): ',train_score)
print('Test Score (R-Squared)',test_score)
```

```python
from sklearn.metrics import mean_squared_error
train_predictions = model.predict(X_train)
train_RMSE=mean_squared_error(Y_train,train_predictions)**0.2
test_predictions = model.predict(X_test)
test_RMSE=mean_squared_error(Y_test,test_predictions)**0.2
print('Train RMSE : ',train_RMSE)
print('Test RMSE : ',test_RMSE)
```
